```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import seaborn as sns
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
train.fillna(0, inplace = True)
train.describe()
corr = train.corr().tail(1)
#np.argsort(df.loc['a'])
corr = corr.sort_values(by = 'SalePrice',axis=1,ascending=False)
corr.T
```

```python

features = ['Lot Frontage', 'Lot Area', 'Garage Area',\
            'Garage Cars', 'Garage Yr Blt', 'Year Remod/Add',\
            'Fireplaces', 'Gr Liv Area', 'Overall Qual',\
            'Year Built', 'Overall Cond', 'Misc Val',\
            'Lot Frontage',  '1st Flr SF', 'Mas Vnr Area',\
            '2nd Flr SF', 'BsmtFin SF 1', 'BsmtFin SF 2']

X = train[features]
y = train['SalePrice']
#poly = PolynomialFeatures(include_bias=False)
#X_poly = poly.fit_transform(X)
train['Sale Type']
X.head()
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 120, random_state = 42)
# Train the model on training data
rf_fitted = rf.fit(X, y)
rf_predict = rf.predict(X)
```

```python
print(cross_val_score(rf_fitted, X, y, cv=5).mean())
rf_rmse = mean_squared_error(train['SalePrice'], rf_predict)
print(rf_rmse ** 0.5)
```

```python
lr = LinearRegression()
lr_fitted = lr.fit(X, y)
print(cross_val_score(lr_fitted, X, y, cv=5).mean())
print(lr.score(X, y))
lr_predict = lr.predict(X)
lr_rmse = mean_squared_error(train['SalePrice'], lr_predict)
print(lr_rmse ** 0.5)

#cross_val_score(lr, X_poly, y).mean()
```

```python
test.fillna(0, inplace = True)
test['price'] = lr.predict(test[features])
```

```python
ids = test['Id']
price = test['price']
df = pd.DataFrame({'Id': ids, 'SalePrice':price}) 
df.sort_values('Id', inplace=True)
```

```python
df.to_csv('./test_submission.csv', index=False)
```
